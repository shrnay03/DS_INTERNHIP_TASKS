```python
import seaborn as sns
import matplotlib.pyplot 
import numpy as np
```

```python
grouped_dataset = dataset.groupby('Continent').sum(numeric_only=True).sort_values(by="2022 Population",ascending=False)
matplotlib.pyplot.figure(figsize=(20,8))
matplotlib.pyplot.bar(grouped_dataset.index, grouped_dataset['2022 Population'],color="pink",edgecolor='blue')
matplotlib.pyplot.xlabel('Continent',  fontsize=22)
matplotlib.pyplot.ylabel('Population(2022)',  fontsize=22)
matplotlib.pyplot.title('Bar-graph Popalation Distribution by Continent 2022',  fontsize=22)
matplotlib.pyplot.xticks(rotation=0, fontsize=18)


for x,y in enumerate(grouped_dataset['2022 Population']):
    matplotlib.pyplot.text(x, y + 200000, str(y), ha='center', va='bottom')

matplotlib.pyplot.show
```

```python
grouped_dataset = dataset.groupby('Continent').sum(numeric_only=True).sort_values(by="2022 Population",ascending=False)
labels = grouped_dataset.index.tolist()
sizes = grouped_dataset["2022 Population"].tolist()
matplotlib.pyplot.figure(figsize=(12,8))
matplotlib.pyplot.pie(sizes,labels=labels, autopct='%1.2f%%',startangle=180)
matplotlib.pyplot.axis('equal')
matplotlib.pyplot.title("Pie chart for Population Distribution by Continent Percentage(2022)", fontsize=22)
matplotlib.pyplot.show()
```
